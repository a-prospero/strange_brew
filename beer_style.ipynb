```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = "Data/style_brewer_data.xlsx"
beer_styles = pd.read_excel(data)
```

```python
df = pd.DataFrame(beer_styles)
df
```

```python
#Eliminate fluff
full_data = df[['Name', 'Style', 'Category', 'Brewer', 'Address', 'State', 'Country', 'Coordinates']]
full_data

full_data = pd.DataFrame(full_data)
full_data
```

```python
#Rename Columns
clean_data = full_data.rename(columns={'Name' :'Beer Name', 'Style':'Beer Style', 'Category':'Category', 'Brewer':'Brewery', 'Address' : 'Address', 'State':'State',
                        'Country':'Country', 'Coordinates': 'Coordinates'})
clean_data
```

```python
clean_data = pd.DataFrame(clean_data)
clean_data

clean_data.to_csv('beer_style.csv')
```

```python
beer_styles = clean_data['Beer Style'].unique()
beer_styles
```

```python
nation_beer_style_count = clean_data.value_counts(clean_data['Beer Style'])
nation_beer_style_count.head(10)
```

```python
nation_beer_style_count
```

```python
nation_beer_style_count.columns = ['Count']
nation_beer_style_count.head()
```

```python
nation_beer_style_count = pd.DataFrame(nation_beer_style_count)
nation_beer_style_count.to_excel('nationwide.xlsx')
```

```python
USA = clean_data.loc[clean_data['Country']=='United States']
USA

USA.to_csv('beer_style.csv')
```

```python
state_dictionary = {
    'CA': 'California', 
    'DC': 'Washington, DC',
    'District of Columbia' : 'Washington, DC',
    'KS': 'Kansas',
    'MA': 'Massachusetts',
    'MI': 'Michigan',
    'MO': 'Missouri',
    'NC': 'North Carolina',
    'NJ': 'New Jersey',
    'NY': 'New York',
    'OH': 'Ohio',
    'Or': 'Oregon',
    'WI': 'Wisconsin'    
    
}
```

```python
USA['State'] = USA['State'].map(lambda x: state_dictionary[x] if x in state_dictionary.keys() else x)
```

```python
state_beer = USA.groupby('State')
state_beer.head()
```

```python
states_count = state_beer['Beer Style'].value_counts()
```

```python
# .index.unique()
```

```python
# beer_count = states_count.value_counts()
```

```python
states_count =pd.DataFrame(states_count)
states_count.rename(columns = {'State': 'State', 'Beer Style': 'Beer Style', 'Beer Style': 'Count'})
```

```python
state_beer = states_count.groupby('State').head(1)
state_beer
```

```python
state_beer.rename(columns = {'State': 'State', 'Beer Style': 'Beer Style', 'Beer Style': 'Count'})
```

```python
state_beer = pd.DataFrame(state_beer)
state_beer.to_csv('state_beer.csv')
```

```python
nations_beer_style = pd.read_excel('nationwide.xlsx')
nations_beer_style
```

```python
x_axis = np.arange(len(nations_beer_style))
tick_locations = [value+0.4 for value in x_axis]
```

```python
plt.figure(figsize=(20,3))
plt.bar(x_axis, nations_beer_style["Count"], color='g', alpha=0.5, align="edge")
plt.xticks(tick_locations, nations_beer_style["Beer Style"], rotation="vertical")
```
