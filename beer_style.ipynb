```python
import pandas as pd
import numpy as np

data = "Data/style_brewer_data.xlsx"
beer_styles = pd.read_excel(data)
```

```python
df = pd.DataFrame(beer_styles)
df
```

```python
#Rename Columns
df = df.rename(columns={'Name' :'Beer Name', 'Style':'Beer Style', 'Category':'Category', 'Brewer':'Brewery', 'Address' : 'Address', 'State':'State',
                        'Country':'Country', 'Coordinates': 'Coordinates'})
```

```python
#Eliminate fluff
full_data = df[['Beer Name', 'Beer Style', 'Category', 'Brewery', 'Address', 'State', 'Country', 'Coordinates']]
full_data

full_data = pd.DataFrame(full_data)
full_data
```

```python
beer_style_states = full_data.groupby('Beer Style')['State'].describe()
print(beer_style_states)
```
